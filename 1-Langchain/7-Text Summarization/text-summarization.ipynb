## Prompt Template Text Summarization


```python
from langchain.chains import LLMChain
from langchain import PromptTemplate

genericTemplate = """
Write a summary of the following content:
Content: {content}
Translate the precise summary to {language}

"""

prompt = PromptTemplate(
    input_variables = ['content', 'language'],
    template = genericTemplate
)
prompt
```

```python
complete_prompt = prompt.format(content=content, language="English")
```

```python
llm.get_num_tokens(complete_prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
summary = llm_chain.run({'content': content, 'language': 'English'})
summary
```

## StuffDocument Chain Text Summarization

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Attention.pdf")
docs = loader.load_and_split()
docs
```

```python
template=""" Write a concise an short summary of the following speech,
Speech:{text}

"""

prompt = PromptTemplate(
    input_variables = ['text'],
    template = template
)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain = load_summarize_chain(llm, chain_type='stuff', prompt=prompt, verbose=True)
output = chain.run(docs)
output
```

### Map Reduce

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
documents = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=100).split_documents(docs)
documents
```

```python
len(documents)
```

```python
chunks_prompt = """
Please summarize the bellow speech:
Speech: {text}
Summary:
"""
map_prompt_template = PromptTemplate(input_variables=['text'],
                                     template=chunks_prompt)
```

```python
final_prompt="""
Provide the final summary of the entire speech with these important points.
Add a Motivation Title, Start the precise summary with an introduction and provide the summary in 
number points for the speech.
Speech:{text}

"""
final_prompt_template = PromptTemplate(input_variables=['text'], template=final_prompt)
```

```python
summary_chain = load_summarize_chain(
    llm=llm,
    chain_type="map_reduce",
    map_prompt=map_prompt_template,
    combine_prompt=final_prompt_template,
    verbose=True
)

output = summary_chain.run(documents)
output
```

### Refine Chain

```python
chain = load_summarize_chain(
    llm=llm,
    chain_type="refine",
    verbose=True
)

output_summary = chain.run(documents)
```
